## 1. Given X be a discrete random variable with the following PMF

$$P_X(x) = \begin{cases}
0.1 & \text{for } x = 0.2 \\
0.2 & \text{for } x = 0.4 \\
0.2 & \text{for } x = 0.5 \\
0.3 & \text{for } x = 0.8 \\
0.2 & \text{for } x = 1 \\
0 & \text{otherwise}
\end{cases}$$

### 1. Find the range RX of the random variable X.
### 2. Find P(X ≤ 0.5)
### 3. Find P(0.25&lt;X&lt;0.75)
### 4. P(X = 0.2|X&lt;0.6)

**Solution:**

1. **Find the range RX of the random variable X:**

   The range of a random variable is the set of all possible values it can take. From the given probability mass function (PMF), we can see that X can take the values 0.2, 0.4, 0.5, 0.8, and 1. So, the range of X, RX, is {0.2, 0.4, 0.5, 0.8, 1}.


2. **Find P(X ≤ 0.5):**

   To find $P(X \leq 0.5)$, you need to sum the probabilities of X being less than or equal to 0.5. 

   $P(X \leq 0.5) = P(X = 0.2) + P(X = 0.4) + P(X = 0.5) = 0.1 + 0.2 + 0.2 = 0.5$

   So, $P(X \leq 0.5) = 0.5$.



3. **Find P(0.25 < X < 0.75):**

   To find $P(0.25 < X < 0.75)$, you need to sum the probabilities of X falling in the interval (0.25, 0.75).

   $P(0.25 < X < 0.75) = P(X = 0.4) + P(X = 0.5) = 0.2 + 0.2 = 0.4$

   So, $P(0.25 < X < 0.75) = 0.4$.


4. **Find P(X = 0.2 | X < 0.6):**

   This is a conditional probability. It represents the probability that X is equal to 0.2 given that X is less than 0.6. To calculate it, we need to consider the probability of X being 0.2 and X being less than 0.6.

   $P(X = 0.2 | X < 0.6) = \frac{P(X = 0.2 \text{ and } X < 0.6)}{P(X < 0.6)}$

   First, calculate the probability in the numerator:

   $P(X = 0.2 \text{ and } X < 0.6) = P(X = 0.2) = 0.1$

   Next, calculate the probability in the denominator:

   $P(X < 0.6) = P(X = 0.2) + P(X = 0.4) + P(X = 0.5) = 0.1 + 0.2 + 0.2 = 0.5$

   Now, calculate the conditional probability:

   $P(X = 0.2 | X < 0.6) = \frac{0.1}{0.5} = 0.2$

   So, $P(X = 0.2 | X < 0.6) = 0.2$.

## 7. Assuming X is a continuous random variable with PDF



$$f_X(x) = \begin{cases}
cx^2 &  |x| \leq 1 \\
0 & \text{otherwise}
\end{cases}$$

## 1. Find EX and Var(X).
## 2. Find P(X ≥ 1/2).

**Solution:**

We need to ensure that the PDF integrates to 1 over the entire support of the random variable $X$. The support in this case is $-1 \leq x \leq 1$.

   $$1 = \int_{-1}^{1} c \cdot x^2 \, dx\$$

   Calculate the integral:

   $1 = c \cdot \left[\frac{x^3}{3}\right]_{-1}^{1} = c \cdot \left[\frac{1^3}{3} - \frac{(-1)^3}{3}\right] = c \cdot \left[\frac{1}{3} + \frac{1}{3}\right] = c \cdot \frac{2}{3}$

   To make the integral equal to 1, we solve for \(c\):

   $c = \frac{3}{2}$

1. **Finding $E[X]$ (Expected Value) and $Var(X)$ (Variance):**

   We can find the expected value and variance of $X$ using the following formulas:

   $E[X] = \int_{-\infty}^{\infty} x \cdot f_X(x) \, dx$
   
   $Var(X) = E[X^2] - (E[X])^2$

   For our PDF $f_X(x) = \frac{3}{2}x^2$, we can calculate:

   $E[X] = \int_{-1}^{1} x \cdot \frac{3}{2}x^2 \, dx$
   

   Calculate the integral:

   $E[X] = \frac{3}{2} \cdot \int_{-1}^{1} x^3 \, dx = \frac{3}{2} \cdot \left[\frac{x^4}{4}\right]_{-1}^{1} = \frac{3}{2} \cdot \left[\frac{1^4}{4} - \frac{(-1)^4}{4}\right] = \frac{3}{2} \cdot \left[\frac{1}{4} + \frac{1}{4}\right] = \frac{3}{2} \cdot \frac{1}{2} = \frac{3}{4}$


   Now, let's find \(E[X^2]\):

   $E[X^2] = \int_{-1}^{1} x^2 \cdot \frac{3}{2}x^2 \, dx$

   Calculate the integral:

   $E[X^2] = \frac{3}{2} \cdot \int_{-1}^{1} x^4 \, dx = \frac{3}{2} \cdot \left[\frac{x^5}{5}\right]_{-1}^{1} = \frac{3}{2} \cdot \left[\frac{1^5}{5} - \frac{(-1)^5}{5}\right] = \frac{3}{2} \cdot \left[\frac{1}{5} + \frac{1}{5}\right] = \frac{3}{2} \cdot \frac{2}{5} = \frac{3}{5}$

   Now, we can find \(Var(X)\):

   $Var(X) = E[X^2] - (E[X])^2 = \frac{3}{5} - \left(\frac{3}{4}\right)^2 = \frac{3}{5} - \frac{9}{16} = \frac{48}{80} - \frac{45}{80} = \frac{3}{80}$


2. **Finding $P(X \geq 1/2)$:**

   To find $P(X \geq 1/2)$, we need to calculate the area under the PDF $f_X(x)$ for $x \geq 1/2$. This can be done by integrating the PDF over the interval $[1/2, 1]$:

   $P(X \geq 1/2) = \int_{1/2}^{1} \frac{3}{2}x^2 \, dx$

   Calculate the integral:

   $P(X \geq 1/2) = \frac{3}{2} \cdot \int_{1/2}^{1} x^2 \, dx$

   $P(X \geq 1/2) = \frac{3}{2} \cdot \left[\frac{x^3}{3}\right]_{1/2}^{1} = \frac{3}{2} \cdot \left[\frac{1^3}{3} - \frac{(1/2)^3}{3}\right] = \frac{3}{2} \cdot \left[\frac{1}{3} - \frac{1}{24}\right] = \frac{3}{2} \cdot \left[\frac{8}{24} - \frac{1}{24}\right] = \frac{3}{2} \cdot \frac{7}{24} = \frac{21}{48} = \frac{7}{16}$

So, $P(X \geq 1/2) = \frac{7}{16}$.

# 8. If X is a continuous random variable with pdf



$$f_X(x) = \begin{cases}
4x^3 & \text{for } 0 < x \leq 1 \\
0 & \text{otherwise}

## Find  $P(X≤2/3∣X>1/3)$.

**Solution:**

$P(X \leq 2/3 | X > 1/3) = \frac{P(1/3 < X \leq 2/3)}{P(X > 1/3)}$


$= \frac{\int_{1/3}^{2/3} \frac{4x^3}{3} \, dx}{\int_{1/3}^1 \frac{4x^3}{3} \, dx}$


$= \frac{\frac{3}{16} \int_{1/3}^{2/3} x^3 \, dx}{\frac{3}{80} \int_{1/3}^1 x^3 \, dx}$


$= \frac{\int_{1/3}^{2/3} x^3 \, dx}{\int_{1/3}^1 x^3 \, dx}$


$= \frac{\left[\frac{x^4}{4}\right]_{1/3}^{2/3}}{\left[\frac{x^4}{4}\right]_{1/3}^1}$


$= \frac{\left(\frac{(2/3)^4}{4}\right) - \left(\frac{(1/3)^4}{4}\right)}{\left(\frac{1^4}{4}\right) - \left(\frac{(1/3)^4}{4}\right)}$


$= \frac{\frac{16}{81} - \frac{1}{81}}{\frac{4}{4} - \frac{1}{81}}$


$= \frac{\frac{15}{81}}{\frac{80}{81}}$


$= \frac{15}{80} = \frac{3}{16}$
